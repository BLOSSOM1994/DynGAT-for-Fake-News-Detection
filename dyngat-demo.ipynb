```python
import pandas as pd

# بارگذاری داده‌ها
train_df = pd.read_csv('/kaggle/input/deep-fake-dataset/train.csv', delimiter=',', nrows = 5000)

# نگاهی به اولین سطرها
train_df.head()
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# بارگذاری کتابخانه‌های مورد نیاز
nltk.download('punkt')
nltk.download('stopwords')
```

```python

# تعریف توکن‌سازی و پیش‌پردازش متن
def preprocess_text(text):
    # تبدیل متن به حروف کوچک
    text = text.lower()
    
    # حذف علائم نگارشی
    text = re.sub(r'[^\w\s]', '', text)
    
    # توکن‌سازی (تبدیل متن به لیستی از کلمات)
    tokens = word_tokenize(text)
    
    # حذف کلمات بی‌معنی (Stop words)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    # بازگرداندن متن پردازش شده
    return ' '.join(tokens)

# پیش‌پردازش تمام متون در DataFrame
train_df['processed_text'] = train_df['text'].apply(preprocess_text)

# نمایش نمونه‌ای از داده‌های پردازش‌شده
train_df[['text', 'processed_text']].head(3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import torch

# استفاده از TF-IDF برای نمایش ویژگی‌های متنی
vectorizer = TfidfVectorizer(max_features=5000)
X_text = vectorizer.fit_transform(train_df['processed_text']).toarray()

# تبدیل به Tensor
text_tensor = torch.tensor(X_text, dtype=torch.float)

# نمایش شکل داده‌ها
print(f"Text features shape: {text_tensor.shape}")
# ورودی متن به مدل به همراه داده‌های گراف
text_input = text_tensor
```

```python
from sklearn.preprocessing import LabelEncoder

# پر کردن مقادیر گمشده
train_df['screen_name'] = train_df['screen_name'].fillna('Unknown')
train_df['account.type'] = train_df['account.type'].fillna('Unknown')

# تبدیل داده‌های متنی به عددی (Label Encoding)
label_encoder = LabelEncoder()
train_df['screen_name'] = label_encoder.fit_transform(train_df['screen_name'])
train_df['account.type'] = label_encoder.fit_transform(train_df['account.type'])


# Replace class_type values to 'fake' and 'real'
train_df['class_type'] = train_df['class_type'].map({
    'human': 0, 
    'rnn': 1, 
    'others': 1, 
    'gpt2': 1
}).astype(int)
```

```python
train_df['user_id'] = pd.to_numeric(train_df['user_id'], errors='coerce')
train_df['user_id'] = train_df['user_id'].fillna(0).astype(int)  # Optionally replace NaNs with 0

train_df['status_id'] = pd.to_numeric(train_df['status_id'], errors='coerce')
train_df['status_id'] = train_df['status_id'].fillna(0).astype(int)  # Optionally replace NaNs with 0
```

```python
# تبدیل شناسه‌های گره‌ها به int یا str
train_df['user_id'] = train_df['user_id'].astype(int)
train_df['status_id'] = train_df['status_id'].astype(int)
```

```python
import matplotlib.pyplot as plt

# Assuming train_df is already processed with encoded features
def plotCorrelationMatrix(df, graphWidth):
    filename = df.dataframeName
    df = df.dropna(axis=1)  # drop columns with NaN
    df = df[[col for col in df if df[col].nunique() > 1]]  # keep columns with more than 1 unique value
    
    if df.shape[1] < 2:
        print(f'No correlation plots shown: The number of non-NaN or constant columns ({df.shape[1]}) is less than 2')
        return
    
    # Compute correlation matrix
    corr = df.corr()
    
    # Plot the correlation matrix
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum=1)
    
    # Customize the ticks and labels
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()  # Move x-axis labels to the bottom
    plt.colorbar(corrMat)
    plt.title(f'Correlation Matrix for {filename}', fontsize=15)
    
    # Show the plot
    plt.show()

# Call the function with the dataframe
plotCorrelationMatrix(train_df, graphWidth=10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# تعریف تابع برای نمایش توزیع داده‌ها
def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique()
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 50]] # برای نمایش، فقط ستون‌هایی که بین 1 تا 50 مقدار منحصر به فرد دارند انتخاب می‌شوند
    nRow, nCol = df.shape
    columnNames = list(df)
    nGraphRow = (nCol + nGraphPerRow - 1) // nGraphPerRow  # تعداد ردیف‌های گراف
    plt.figure(num=None, figsize=(6 * nGraphPerRow, 8 * nGraphRow), dpi=80, facecolor='w', edgecolor='k')
    
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        
        # اگر ستون عددی است، هیستوگرام رسم می‌شود
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar()
        else:
            columnDf.hist()
        
        plt.ylabel('counts')
        plt.xticks(rotation=90)  # چرخاندن برچسب‌ها برای خوانایی بیشتر
        plt.title(f'{columnNames[i]} (column {i})')
    
    plt.tight_layout(pad=1.0, w_pad=1.0, h_pad=1.0)
    plt.show()

# فراخوانی تابع برای داده‌های train_df
plotPerColumnDistribution(train_df, 8, 4)
```

```python
import networkx as nx
import random

# ایجاد گراف خالی
G = nx.Graph()

# ایجاد مجموعه‌ای برای ذخیره کاربر و مقاله‌های یکتا
unique_users = set()
unique_news = set()

# افزودن گره‌ها و لبه‌ها به گراف
for index, row in train_df.iterrows():
    user_node = row['user_id']
    news_node = row['status_id']
    
    # فقط گره‌های یکتا را به گراف اضافه می‌کنیم
    if user_node not in unique_users:
        G.add_node(user_node, type='user')
        unique_users.add(user_node)
    
    if news_node not in unique_news:
        G.add_node(news_node, type='news')
        unique_news.add(news_node)
    
    # افزودن لبه‌ها (ارتباطات)
    G.add_edge(user_node, news_node)

# انتخاب 300 گره تصادفی
selected_nodes = random.sample(list(G.nodes()), 300)  # تبدیل مجموعه به لیست

# انتخاب یال‌های مرتبط با این 300 گره
edges = set()
for node in selected_nodes:
    for neighbor in G.neighbors(node):
        if neighbor in selected_nodes:
            edges.add(tuple(sorted((node, neighbor))))  # جلوگیری از تکرار یال‌ها

# ایجاد گراف جدید با گره‌ها و یال‌های انتخابی
subgraph = G.subgraph(selected_nodes).copy()

# اضافه کردن یال‌های جدید به گراف
subgraph.add_edges_from(edges)

# نمایش تعداد گره‌ها و یال‌ها در گراف جدید
print(f"Nodes: {len(subgraph.nodes)}")
print(f"Edges: {len(subgraph.edges)}")
```

```python
import random
import networkx as nx
import torch
import pandas as pd

# فرض کنید train_df به شکل زیر باشد:
# Columns: ['user_id', 'status_id', 'screen_name', 'account.type', 'class_type', 'text_length']

# ساخت گراف اصلی
G = nx.Graph()

# افزودن گره‌های کاربران
for idx, row in train_df.iterrows():
    G.add_node(row['user_id'], type='user', screen_name=row['screen_name'], account_type=row['account.type'])

# افزودن گره‌های خبرها
for idx, row in train_df.iterrows():
    G.add_node(row['status_id'], type='news', account_type='real' if row['account.type'] == 0 else 'fake', text_length=row['text_length'])

# افزودن یال‌ها بین کاربران و خبرها (ارتباطات)
for idx, row in train_df.iterrows():
    G.add_edge(row['user_id'], row['status_id'], interaction="interaction", weight=random.randint(1, 3))

# انتخاب 300 گره تصادفی از گراف اصلی
selected_nodes = random.sample(list(G.nodes()), min(1000, len(G.nodes())))

# انتخاب یال‌های مرتبط با گره‌های انتخابی
edges = set()
for node in selected_nodes:
    for neighbor in G.neighbors(node):
        if neighbor in selected_nodes:
           edges.add(tuple(sorted((node, neighbor))))  # جلوگیری از تکرار یال‌ها

# ایجاد نگاشت گره‌ها به اندیس‌ها
node_to_index = {node: idx for idx, node in enumerate(selected_nodes)}

# ساخت edge_index
edges = list(edges)
edge_index = torch.tensor([[node_to_index[edge[0]], node_to_index[edge[1]]] for edge in edges], dtype=torch.long).t().contiguous()

# ساخت ویژگی‌های گره‌ها
node_features = []
for node in selected_nodes:
    node_type = G.nodes[node].get('type', 'unknown')  # 'user' یا 'news'
    if node_type == 'user':
        # ویژگی‌های کاربر
        user_data = train_df[train_df['user_id'] == node].iloc[0]
        features = [
            len(G[node]),  # تعداد یال‌ها
            1 if user_data['account.type'] == 'human' else 0,  # نوع حساب
        ]
    elif node_type == 'news':
        # ویژگی‌های خبر
        news_data = train_df[train_df['status_id'] == node].iloc[0]
        features = [
            news_data['text_length'],  # طول متن
            1 if news_data['account.type'] == 0 else 0,  # نوع خبر دهنده
        ]
    else:
        features = [0, 0]  # ویژگی پیش‌فرض برای گره‌های ناشناخته
    node_features.append(features)

# تبدیل ویژگی‌ها به Tensor
node_features = torch.tensor(node_features, dtype=torch.float)

# بررسی خروجی
print("Edge Index:")
print(edge_index.shape)
print("Node Features:")
print(node_features.shape)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GATConv

# 1. Graph Attention Module
class GraphAttentionModule(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(GraphAttentionModule, self).__init__()
        self.gat = GATConv(in_channels, out_channels, heads=1, concat=False)

    def forward(self, x, edge_index):
        x = self.gat(x, edge_index)
        return F.elu(x)  # استفاده از تابع فعال‌سازی ELU
```

```python
# 2. News Article Classifier
class NewsArticleClassifier(nn.Module):
    def __init__(self, input_dim, rnn_hidden_size, cnn_out_channels, num_classes):
        super(NewsArticleClassifier, self).__init__()
        # LSTM برای مدل‌سازی وابستگی‌های زمانی در متن
        self.rnn = nn.LSTM(input_dim, rnn_hidden_size, batch_first=True)
        
        # CNN برای استخراج ویژگی‌های محلی
        self.conv1d = nn.Conv1d(rnn_hidden_size, cnn_out_channels, kernel_size=1)  # ویژگی‌های RNN به CNN داده می‌شود
        
        # مکانیزم توجه برای تمرکز روی بخش‌های مهم
        self.attention = nn.Linear(cnn_out_channels, 1)
        
        # لایه تمام‌متصل برای طبقه‌بندی
        self.fc = nn.Linear(cnn_out_channels, num_classes)  # توجه به ویژگی‌های CNN
    def forward(self, text_input):
        # تغییر شکل ورودی برای LSTM به [batch_size, seq_len, input_size]
        rnn_out, (h_n, c_n) = self.rnn(text_input)  # خروجی RNN
   #     print("RNN output shape:", rnn_out.shape)  # ابعاد خروجی RNN را چاپ می‌کند

        # انتخاب آخرین وضعیت پنهان
       # rnn_out = h_n[-1]  # خروجی آخرین وضعیت پنهان LSTM
     #   print("Hidden state (h_n[-1]) shape:", rnn_out.shape)  # ابعاد آخرین وضعیت پنهان را چاپ می‌کند

        # تغییر ابعاد rnn_out به [batch_size, rnn_hidden_size, seq_len]
        rnn_out = rnn_out.unsqueeze(1)  # تغییر به [batch_size, 1, seq_len]
    #    print("After unsqueeze(1):", rnn_out.shape)
        
        # بعد از unsqueeze، باید ابعاد rnn_out به [batch_size, rnn_hidden_size, seq_len] تغییر داده شود
        # زیرا کانولوشن انتظار دارد که ورودی به شکل [batch_size, channels, seq_len] باشد.
        rnn_out = rnn_out.permute(0, 2, 1)  # تغییر به ابعاد [batch_size, seq_len, rnn_hidden_size]
      #  print("After permute:", rnn_out.shape)
        
        # ویژگی‌های RNN به CNN داده می‌شود
        conv_out = self.conv1d(rnn_out)  # اعمال CNN
        conv_out = F.relu(conv_out)
        conv_out = torch.max(conv_out, dim=2)[0]  # عملیات حداکثر در طول دیمشن توالی
        
        # اعمال مکانیزم توجه به ویژگی‌های CNN
        attention_scores = torch.softmax(self.attention(conv_out), dim=1)
        attention_out = attention_scores * conv_out

        # طبقه‌بندی نهایی با استفاده از ویژگی‌های توجه
        output = self.fc(attention_out)
        return output
```

```python
class GraphAndTextClassifier(nn.Module):
    def __init__(self, graph_in_channels, graph_out_channels, text_input_dim, rnn_hidden_size, cnn_out_channels, num_classes):
        super(GraphAndTextClassifier, self).__init__()
        
        # Graph Attention Module
        self.graph_attention = GraphAttentionModule(graph_in_channels, graph_out_channels)
        
        # News Article Classifier
        self.news_article_classifier = NewsArticleClassifier(text_input_dim, rnn_hidden_size, cnn_out_channels, num_classes)
        
        # لایه تمام‌متصل برای ترکیب ویژگی‌های گراف و متن
        # ابعاد ورودی این لایه برابر با مجموع ویژگی‌های گراف و متن خواهد بود
        self.fc_combined = nn.Linear(graph_out_channels + cnn_out_channels, num_classes)  # ورودی 66 و خروجی num_classes

    def forward(self, node_features, edge_index, text_input):
        # استخراج ویژگی‌های گراف
        graph_features = self.graph_attention(node_features, edge_index)
        
        # استخراج ویژگی‌های متنی
        text_features = self.news_article_classifier(text_input)
        
        # بررسی ابعاد ویژگی‌های گراف و متن
       # print(f"graph_features shape: {graph_features.shape}, text_features shape: {text_features.shape}")
        
        # ترکیب ویژگی‌های گراف و متنی
        combined_features = torch.cat((graph_features, text_features), dim=1)  # ترکیب ویژگی‌ها
        
        # بررسی ابعاد ترکیب شده
       # print(f"combined_features shape: {combined_features.shape}")
        
        # طبقه‌بندی نهایی
        output = self.fc_combined(combined_features)  # پیش‌بینی نهایی
        return output

# تعریف مدل
model = GraphAndTextClassifier(
    graph_in_channels=2,  # تعداد ویژگی‌ها برای گره‌ها
    graph_out_channels=2,  # تعداد ویژگی‌های خروجی گراف (پس از گراف ATT)
    text_input_dim=1987,  # تعداد ویژگی‌های هر مقاله (۱۹۸۷ ویژگی برای هر مقاله)
    rnn_hidden_size=128,  # اندازه مخفی RNN
    cnn_out_channels=2,  # تعداد کانال‌های خروجی CNN
    num_classes=2  # تعداد کلاس‌ها (واقعی یا جعلی)
)

# پیش‌بینی با مدل
output = model(node_features, edge_index, text_input)
print("Model output shape:", output.shape)   # باید خروجی به شکل [300, 2] باشد
```

```python
import torch
import torch.optim as optim

# تعریف تابع از دست دادن و بهینه‌ساز
criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# آموزش مدل بدون استفاده از DataLoader
for epoch in range(200):  # 10 دوره آموزشی
    model.train()  # مدل در حالت آموزش
    optimizer.zero_grad()  # صفر کردن گرادیان‌ها
    
    # پیش‌بینی با مدل
    outputs = model(node_features, edge_index, text_input)
    
    # محاسبه از دست دادن
    loss = criterion(outputs, labels)
    loss.backward()  # محاسبه گرادیان‌ها
    optimizer.step()  # به‌روزرسانی پارامترها
    
    # محاسبه دقت
    _, predicted = torch.max(outputs, 1)
    correct = (predicted == labels).sum().item()
    accuracy = 100 * correct / len(labels)
    
    print(f"Epoch {epoch+1},\t Loss: {loss.item():.4f},\t Accuracy: {accuracy:.2f}%")
```

```python
model.eval()  # مدل در حالت ارزیابی
with torch.no_grad():  # بدون محاسبه گرادیان
    outputs = model(node_features, edge_index, text_input)
    
    # محاسبه دقت
    _, predicted = torch.max(outputs, 1)
    correct = (predicted == labels).sum().item()
    accuracy = 100 * correct / len(labels)
    
    print(f"Test Accuracy: {accuracy:.2f}%")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# پیش‌بینی‌ها و برچسب‌های واقعی برای ارزیابی
all_preds = []
all_labels = []

# ارزیابی مدل
with torch.no_grad():  # بدون محاسبه گرادیان
    for i in range(0, len(labels), 32):  # پردازش داده‌ها به صورت دسته‌ای
        # استخراج بخش‌هایی از داده‌ها
        batch_node_features = node_features
        batch_edge_index = edge_index
        batch_text_input = text_input
        batch_labels = labels
        
        # پیش‌بینی با مدل
        outputs = model(batch_node_features, batch_edge_index, batch_text_input)
        
        # ذخیره پیش‌بینی‌ها و برچسب‌ها
        _, predicted = torch.max(outputs, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(batch_labels.cpu().numpy())

# گزارش طبقه‌بندی
report = classification_report(all_labels, all_preds)
print(report)

# ماتریس اشتباه
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Fake', 'Real'], yticklabels=['Fake', 'Real'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
